```python
import pandas as pd
import numpy as np

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score

from lightgbm import LGBMClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
path = "/content/drive/MyDrive/Data/"

train = pd.read_csv(path + "Train.csv")
test = pd.read_csv(path + "Test.csv")
ss = pd.read_csv(path + "SampleSubmission.csv")
```

```python
ss.head()
```

```python
def feature_engineering(df):

    df = df.copy()

    df["urban_phone"] = ((df["location_type"] == "Urban") &
                         (df["cellphone_access"] == "Yes")).astype(int)

    df["stable_income"] = df["job_type"].isin([
        "Formally employed Private",
        "Formally employed Government",
        "Self employed"
    ]).astype(int)

    df["age_group"] = pd.cut(
        df["age_of_respondent"],
        bins=[0, 18, 25, 35, 50, 100],
        labels=[0, 1, 2, 3, 4]
    ).astype(int)

    df["household_group"] = pd.cut(
        df["household_size"],
        bins=[0, 2, 4, 6, 10, 50],
        labels=[0, 1, 2, 3, 4]
    ).astype(int)

    return df


train = feature_engineering(train)
test = feature_engineering(test)
```

```python
target = "bank_account"
train[target] = train[target].map({"Yes": 1, "No": 0})
```

```python
cat_cols = train.select_dtypes(include="object").columns
cat_cols = cat_cols.drop("uniqueid")

for col in cat_cols:
    freq = train[col].value_counts(normalize=True)

    train[col] = train[col].map(freq)
    test[col] = test[col].map(freq)
```

```python
X = train.drop(["bank_account", "uniqueid"], axis=1)
y = train["bank_account"]

test_ids = test["uniqueid"]
X_test = test.drop(["uniqueid"], axis=1)
```

```python
model = LGBMClassifier(
    n_estimators=800,
    learning_rate=0.03,
    num_leaves=31,
    max_depth=-1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

scores = []
test_preds = np.zeros(len(X_test))

for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):

    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model.fit(X_tr, y_tr)

    val_pred = model.predict(X_val)
    acc = accuracy_score(y_val, val_pred)
    scores.append(acc)

    test_preds += model.predict(X_test) / 5

    print(f"Fold {fold+1}: {acc}")

print("Mean CV Score:", np.mean(scores))
```

```python
submission = pd.DataFrame({
    "uniqueid": test_ids.fillna("").astype(str) + " x " + test["country"].fillna("").astype(str),
    "bank_account": test_preds.round().astype(int)
})

submission.to_csv("submission.csv", index=False)
```
